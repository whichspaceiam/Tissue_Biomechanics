We have a collagen fibril, tail tendon of a mouse tested under tension and until fracture in single ramp test. We have data containing force in micronewton ($\mu N$) and the fibril displacement in micrometers ($\mu m$). Initial gauge lenght of the collagen fibril is 75.8 $\mu m$ while the collagen fibril is assumed to have a cylinderical shape with a uniform cross-section of 186 $nm$ in diameter.

We have to plot the tangent stiffness (in N/m) versus stretch diagram. We know that tangent stiffness refers to stiffness at a specific point on stress-strain curve (it is slope). 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

data = pd.read_csv("force_disp_v0.csv")

gauge_length = 75.8  # initial gauge length (micrometers)
diameter = 186  # fibril diameter (nanometers)
radius = diameter / 2 * 1e-3  # convert radius to micrometers
k_p = 1.38e-23  # Boltzmann constant in J/K
T = 298  # Temperature in Kelvin
A = np.pi * radius**2  # cross-sectional area (micrometers^2)
N_x = 1000  # Assumed number of Kuhn segments (modify as needed) 

data['Stress'] = data['Force_uN'] / A  # Stress = Force / Area
data['Strain'] = data['Disp_um'] / gauge_length  # Strain = Displacement / Initial Length

# Ensure 'Strain' is strictly increasing
data = data.sort_values(by='Strain')  # Sort by strain
data = data.drop_duplicates(subset=['Strain'])  # Remove duplicate strain values

strain = np.linspace(data['Strain'].min(), data['Strain'].max(), 500)
stress = make_interp_spline(data['Strain'].values, data['Stress'].values)(strain)
tangent_stiffness = np.gradient(stress, strain)


plt.figure(figsize=(8, 6))
plt.plot(strain, stress, label="Stress-Strain Curve", color="blue", linewidth=2)
plt.xlabel("Strain", fontsize=12)
plt.ylabel("Stress (μN/μm²)", fontsize=12)
plt.title("Stress-Strain Curve of Collagen Fibril", fontsize=14)
plt.legend(fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data['Stretch'] = (gauge_length + data['Disp_um']) / gauge_length
tangent_stiffness_Nm = tangent_stiffness * 1e3  # (1 μN/μm² = 10³ N/m)

stretch = make_interp_spline(data['Strain'], data['Stretch'])(strain)

plt.figure(figsize=(8, 6))
plt.plot(stretch, tangent_stiffness_Nm, label="Tangent Stiffness (N/m)", color="red", linewidth=2)
plt.xlabel("Stretch", fontsize=12)
plt.ylabel("Tangent Stiffness (N/m)", fontsize=12)
plt.title("Tangent Stiffness vs. Stretch", fontsize=14)
plt.legend(fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data["FJC Force (N)"] = (3 * k_p * T * N_x / gauge_length) * data['Disp_um']
# Compute FJC-based stress
data["FJC Stress (MPa)"] = data["FJC Force (N)"] / A

# Compute Green-Lagrange strain
data["Green-Lagrange Strain"] = 0.5 * (data["Stretch"] ** 2 - 1)
```
